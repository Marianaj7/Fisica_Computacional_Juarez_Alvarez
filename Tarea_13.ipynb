# Ejercicio 1

Resolver oscilador amortiguado: $$m \ddot x=-k x -b \dot x,$$ usando $m=k=b=1$ y $x_0=1$, $\dot x_0=0$. Usen Euler Y Runge-Kutta orden 4, comparen las soluciones, es decir, grafiquen el espacio de fase para cada método y despues juntos. Nota: El método de Runge-Kutta esta dentro de la familia de métodos simplécticos, es decir, preserva el volumen en el espacio de fase, algo particularmente útil cuanto tenemos sistemas que conservan la energia, a diferencia de Euler que no lo hace. Observen que al no haber cantidades conservadas para el oscilador amortiguado, Euler funcione mucho mejor que en el otro caso, pues el oscilador amortiguado no conserva el volumen del espacio de fase.

```julia
#Se define la función que rige un oscilador armónico amortiguado.
function D(x,t)
    return [x[2], -1.0*x[1]-1.0*x[2]]
end
```

```julia
#Tanto el código para el método de Euler, como el de Runge Kutta son tal cual fueron presentados en la tarea 12.
function euler(f,list,x0)
     x = x0
     h = list[2]-list[1]
     listx=[]
     push!(listx,x)
     for i in 2:length(list)
        t = i*h
        x = x + f(x,t)*h
        push!(listx,x) 
     end
     return listx
end
```

```julia
function runge_kutta2(f,list,x0)
     x = x0
     h = list[2]-list[1]
     listx=[]
     push!(listx,x)
     for i in 2:length(list)
        t = i*h
        k1 = f(x,t);
        k2 = f(x+(h/2)*k1,t+(h/2));
        k3 = f(x+(h/2)*k2, t+(h/2));
        k4 = f(x+h*k3, t+h);
        x = x+(h/6)*(k1+2*k2+2*k3+k4);
        push!(listx,x) 
     end
     return listx
end
```

```julia
#Se define el intervalo list en el cual se evaluará el método de Euler o Runge Kutta, según cual sea el caso. En EE6 y RK7 se guardarán ambos métodos evaluados en la función antes definida, list, y finalmente en la condición inicial.
list=0:0.01:100;
EE6=euler(D,list,[0,1]);
RK7=runge_kutta2(D,list,[0,1]);
```

```julia
list1=map(x->x[1],EE6);
list2=map(x->x[2],EE6);
list3=map(x->x[1],RK7);
list4=map(x->x[2],RK7);
```

```julia
list1
```

```julia
#Se grafica para el método de Euler.
plot(list1,list2,label="Método de Euler");
legend()
```

```julia
#Se grafica para el método de Runge Kutta
plot(list3,list4,label="Método de Runge Kutta");
legend()
```

```julia
#Finalmente, se grafican ambos métodos.
plot(list1,list2,label="Método de Euler");
plot(list3,list4,label="Método de Runge Kutta");
legend()
```

Ajuste el rango de la gráfica para que se centre cerca del cero para que se puedan observar los dos tipos de movimiento presentes. Coménte e interprete las soluciones.

```julia
#Se define la función que rige el comportamiento de un péndulo simple.
function P(x,t)
    return [x[2], -1.0*sin(x[1])]
end
```

```julia
#El proceso que se realiza ahora, es completamente análogo al anterior. Sin embargo, aquí el método toma diferentes condiciones iniciales definidas en forma manual. 
list=0:0.001:10;
RK8=runge_kutta2(P,list,[-0.1,0.1]);
RK9=runge_kutta2(P,list,[0,0.2]);
RK10=runge_kutta2(P,list,[0,0.3]);
RK11=runge_kutta2(P,list,[0,0.02]);
RK12=runge_kutta2(P,list,[0,0.06]);
RK13=runge_kutta2(P,list,[0,1.0]);
RK14=runge_kutta2(P,list,[0,1.004]);
RK15=runge_kutta2(P,list,[0,1.008]);
RK16=runge_kutta2(P,list,[0,1.000004]);
RK17=runge_kutta2(P,list,[0,2.0]);
RK18=runge_kutta2(P,list,[0,2.004]);
RK19=runge_kutta2(P,list,[0,3.008]);
RK20=runge_kutta2(P,list,[0,4.000004]);
```

```julia
list5=map(x->x[1],RK8);
list6=map(x->x[2],RK8);
list7=map(x->x[1],RK9);
list8=map(x->x[2],RK9);
list9=map(x->x[1],RK10);
list10=map(x->x[2],RK10);
list11=map(x->x[1],RK11);
list12=map(x->x[2],RK11);
list13=map(x->x[1],RK12);
list14=map(x->x[2],RK12);
list15=map(x->x[1],RK13);
list16=map(x->x[2],RK13);
list17=map(x->x[1],RK14);
list18=map(x->x[2],RK14);
list19=map(x->x[1],RK15);
list20=map(x->x[2],RK15);
list21=map(x->x[1],RK16);
list22=map(x->x[2],RK16);
list23=map(x->x[1],RK17);
list24=map(x->x[2],RK17);
list25=map(x->x[1],RK18);
list26=map(x->x[2],RK18);
list27=map(x->x[1],RK19);
list28=map(x->x[2],RK19);
list29=map(x->x[1],RK20);
list30=map(x->x[2],RK20);
```

```julia
#A pesar de que tomamos un buen número de condiciones iniciales, nuestro digrama fase no es muy representativo, por ello optamos graficar un sin fin de condiciones iniciales aleatorias.
plot(list5,list6);
plot(list7,list8)
plot(list9,list10)
plot(list11,list12)
plot(list13,list14)
plot(list15,list16)
plot(list17,list18)
plot(list19,list20)
plot(list21,list22)
plot(list23,list24)
plot(list25,list26)
plot(list27,list28)
plot(list29,list30)
legend()
```

```julia
#Nuevamente se define el intervalos a evaluar.
list=0:0.01:10
#Este for iterará 100 veces para obtener las condiciones iniciales aleatorias.
for i in 1:100
    CI=rand(2,1)*9-4.5
    RK21=runge_kutta2(P,list,CI)
    plot(map(x->x[1],RK21),map(x->x[2],RK21))
end
#Se establece el rango en el que será graficado nuestro espacio fase, esto con el fin de poder representar los dos tipos de movimiento presentes.
axes()[:set_xlim]([-5,5])
axes()[:set_ylim]([-5,5])
```

A primera vista es fácil apreciar que la representación de un péndulo simple en el espacio de las fases corresponde a una curva cerrada (en el centro), lo cual indica que el sistema vuelve al mismo estado después de realizar un ciclo completo.
Observamos que la trayectoria en el espacio de las fases es simétrica respecto del eje vertical (eje que representa la velocidad), esto quiere decir que el movimiento del péndulo es el mismo en el sentido a las manecillas del reloj. 
Sin embargo, también se presenta otro tipo de movimiento, el cual está representado por las curvas abiertas. Éstas se generan cuando el péndulo da vueltas completas, o bien, cuando la energía total del péndulo es mayor que el máximo valor de la energía potencial.

# Ejercicio 3

Comprobar numéricamente (con Runge-Kutta orden $4$) que en el sistema: $$\frac{dx}{dt}=x^2-1,$$ 

hay dos puntos fijos, ubicados en $1$ y $-1$, compruebe además que el $1$ es un repulsor y el $-1$ un atractor. Para hacer esto necesitará graficar muchas condiciones iniciales, generelas aleatoriamente y gráfiquelas en el espacio $(x(t),t)$.

```julia
#Se define la nueva función a resolver por el método de Runge Kutta.
function M(x,t)
    return [(x[1]^2)-1,0]
end
```

```julia
list=0:0.01:10
for i in 1:50
S=-2*rand(1)+1   
CI=[S[1],0]
RK22=runge_kutta2(M,list,CI)
   plot(map(M->M[1],RK22))   
end
```

Sin mayor dificultad es claro apreciar que el punto 1 es un punto repulsor, contrario al punto -1 que es atractor.
