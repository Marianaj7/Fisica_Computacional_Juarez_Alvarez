# Oscilador amortiguado

```julia
function D(x,t)
    return [x[2], -1.0*x[1]-1.0*x[2]]
end
```

```julia
function euler(f,list,x0)
     x = x0
     h = list[2]-list[1]
     listx=[]
     push!(listx,x)
     for i in 2:length(list)
        t = i*h
        x = x + f(x,t)*h
        push!(listx,x) 
     end
     return listx
end
```

```julia
function runge_kutta2(f,list,x0)
     x = x0
     h = list[2]-list[1]
     listx=[]
     push!(listx,x)
     for i in 2:length(list)
        t = i*h
        k1 = f(x,t);
        k2 = f(x+(h/2)*k1,t+(h/2));
        k3 = f(x+(h/2)*k2, t+(h/2));
        k4 = f(x+h*k3, t+h);
        x = x+(h/6)*(k1+2*k2+2*k3+k4);
        push!(listx,x) 
     end
     return listx
end
```

```julia
list=0:0.01:100;
EE6=euler(D,list,[0,1]);
RK7=runge_kutta2(D,list,[0,1]);
```

```julia
list1=map(x->x[1],EE6);
list2=map(x->x[2],EE6);
list3=map(x->x[1],RK7);
list4=map(x->x[2],RK7);
```

```julia
list1
```

```julia
plot(list1,list2,label="Método de Euler");
legend()
```

```julia
plot(list3,list4,label="Método de Runge Kutta");
legend()
```

```julia
plot(list1,list2,label="Método de Euler");
plot(list3,list4,label="Método de Runge Kutta");
legend()
```

# Péndulo simple

```julia
function P(x,t)
    return [x[2], -1.0*sin(x[1])]
end
```

```julia
list=0:0.001:10;
RK8=runge_kutta2(P,list,[0,0.1]);
RK9=runge_kutta2(P,list,[0,0.2]);
RK10=runge_kutta2(P,list,[0,0.3]);
RK11=runge_kutta2(P,list,[0,0.02]);
RK12=runge_kutta2(P,list,[0,0.06]);
```

```julia
list5=map(x->x[1],RK8);
list6=map(x->x[2],RK8);
list7=map(x->x[1],RK9);
list8=map(x->x[2],RK9);
list9=map(x->x[1],RK10);
list10=map(x->x[2],RK10);
list11=map(x->x[1],RK11);
list12=map(x->x[2],RK11);
list13=map(x->x[1],RK12);
list14=map(x->x[2],RK12);
```

```julia
plot(list5,list6);
plot(list7,list8)
plot(list9,list10)
plot(list11,list12)
plot(list13,list14)
legend()
```
