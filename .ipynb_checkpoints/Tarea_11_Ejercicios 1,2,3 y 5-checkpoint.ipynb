# Ejercicio 1

```julia
#Se instalan las librerías a utilizar tanto en este ejercicio como el en ejercicio 4 y 5.
Pkg.add("PyPlot")
using PyPlot
using Plots
```

```julia
#Se deine la función de dos variables a resolver por el método de Euler.
function f(x,t)
    return -2.5x
end
```

```julia
#El siguiente código se implementa para resolver una función por el método de Euler, el usuario ingresa la función, la condición inicial, los límites inferior y superior del intervalo de interés y finalmente el ancho de los subintervalos.
function metodo_euler(f,x0,t0,t,h)
    listt=t0:h:t
    listx=[x0]
    for i in 1:length(listt)-1
        x0=x0+f(Float64(x0),listt[i])*h
        push!(listx,x0)
    end
    return listt,listx
end
```

```julia
#Se define función para la solución exacta, ésta sólo depende de t.
function Solucion_exacta(t)
    return e^(-2.5*t)
end
```

```julia
#Ahora se considera la primera condición inicial x(0)=1, se toman los límites del intervalo [0,8] y finalmente se varía el valor de h, según cual sea el caso.
P1=metodo_euler(f,1.0,0.0,8.0,1.0)  
P2=metodo_euler(f,1.0,0.0,8.0,0.5) 
P3=metodo_euler(f,1.0,0.0,8.0,0.2)  
P4=metodo_euler(f,1.0,0.0,8.0,0.01) 
```

```julia
plot!(P2)
plot!(P3)
plot!(P4)
plot!(linspace(0,10,50),Solucion_exacta,label="Solución exacta")
```

```julia
#Se hace aparte la representación gráfica para h=1
plot!(P1)
plot!(linspace(0,10,50),Solucion_exacta,label="Solución exacta")
```

```julia
#Análogamente para la condición inicial x(0)=2
P5=metodo_euler(f,2.0,0.0,8.0,1.0)
P6=metodo_euler(f,2.0,0.0,8.0,0.5)
P7=metodo_euler(f,2.0,0.0,8.0,0.2)
P8=metodo_euler(f,2.0,0.0,8.0,0.01)
```

```julia
plot!(linspace(0,10,50),Solucion_exacta,label="Solución exacta")
plot!(P6)
plot!(P7)
plot!(P8)
```

```julia
plot!(P5)
plot!(linspace(0,10,10),Solucion_exacta,label="Solución exacta")
```

# Ejercicio 2

¿Qué sistema físico está modelado por $\ddot{x} =-kx$? Escriba la solución exacta en terminos de las condiciones iniciales $x(0)$ y $x'(0)$.

Corresponde a un oscilador armónico, el cual es un sistema que, desplazado de su posición de equilibrio, experimenta una fuerza de restitución, que es proporcional al desplazamiento $x$, $k$ corresponde a la constante de elasticidad y el signo negativo indica que cuando $x$ es positiva la fuerza está dirigida hacia las $x$ negativas.

Ahora, la solución de esta ecuación diferencial son funciones cuya segunda derivada es la misma función con signo negativo, es decir: senos y cosenos:

$$x(t)=cos(\sqrt{k}t)$$ y  $$x(t)=sin(\sqrt{k}t)$$

Por lo tanto, la solución general es una combinación lineal de senos y cosenos:

$$x(t)=Acos(\sqrt{k}t)+Bsin(\sqrt{k}t)$$

Entonces, tomando las siguientes condiciones iniciales:


La posición inicial $x(0)=x_0$

La velocidad inicial $\dot{x}(0)=v_0$

La solución será: $$x(t)=x_0cos(\sqrt{k}t)+\frac{v_0}{\sqrt{k}}sin(\sqrt{k}t)$$

# Ejercicio 5

```julia
#Se define la función de dos variables a resolver por el método de Euler.
function g(x,t)
    return (2x-18t)/(1+t)
end
```

```julia
#Se define la función que corresponde a la solución exacta.
function solucion(t)
    return 4+8*t-5t^2
end
```

```julia
#Posteriormente se aplica el método de Euler para resolver la ecuación diferencial. Nótese que la condición inicial se obtuvo por medio de la solución exacta, concluyendo que x(0)=4, mientras que los valores tanto del intervalo como de h son valores arbitrarios.
G=metodo_euler(g,4.0,0.0,3,0.001)
```

```julia
#Se hace la representación gráfica tanto de la solución por el método de Euler, como de la solución exacta, para hacer una comparación.
plot!(G,label="Solución por euler")
plot!(linspace(1,10,50),solucion, label="Solución exacta")
```
