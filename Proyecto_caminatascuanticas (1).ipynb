## 1. Caso clásico

El modelo básico de las caminatas aleatorias es el movimiento de una partícula (llamado caminante) sobre puntos discretos distribuidos en una línea sin restricciones. El sentido del movimiento del caminante (izquierda o derecha) depende de un
sistema bivaluado (una moneda), cuyos valores, para cada paso, dependen de la probabilidad. Si el resultado del voltado es sol(o bien 1, en nuestro caso), entonces el caminante dará un paso a la derecha, mientras que si cae águila (-1) el caminante hará lo contrario, y dará un paso hacia la izqierda. Cabe mencionar que durante todo el problema, el caminante partirá de un punto de inicio, cuyo valor será 0.  Después de muchos volados (digamos, cien), uno puede hacer varias preguntas interesantes, por ejemplo: ¿con qué frecuencia el caminante cae en la posición de inicio?

```julia
#La rutina que se presenta a continuación es una recopilación  de diversas funciones. El input (n) corresponde al número de volados que se desea realizar.
function posicion_final(n)
    b=rand(-1:2:1,n)     #Usamos la función rand para generar valores aleatorios, éstos sólo podrán ser 1 (sol) y -1 (águila).
    p=0
    for i in 1:length(b)
    p=b[i]+p  #Se implementa un for para, una vez generados todos los valores de los volados, obtener la posición final del caminante, esto es simplemente la suma de todos ellos.
end
return p  #Finalmente, el valor p corresponde a la posición final del caminante después de n pasos.
end
```

```julia
#Sin embargo, a nosotros nos interesa hacer un análisis para  K número de caminantes. Ahora le pedimos al usuario tanto el número de volados "n", como el de caminantes "k"
function pf_caminantes(n,k) 
    listf=zeros(k)  # Construimos una lista con k entradas de ceros 
    for i in 1:k  #Este ciclo for se usa para encontrar la posición final de los k caminantes haciendo uso de la función de posicion_final (presentada anteriormente)
        d=posicion_final(n)
        listf[i]=d  
        end 
    return listf # Se imprime la lista con la posición final de los k caminantes.
    end
```

Ahora calculamos la probabilidad para obtener la distribución.

```julia
function proba(n,k)
    A=hist(pf_caminantes(n,k))
    B=A[2]
    x=zeros(length(B))
    for i in 1:length(B)
PA= B[i]/k
        x[i]=PA
        end 
    return x
    end 
```

# 2. Caso cuántico

En este modelo participan dos elementos: el caminante y la moneda (la misma idea que con la caminata aleatoria). Ambos elementos son sistemas físicos cuyo comportamiento se modela y cuantifica mediante los principios y leyes de la mecánica cuántica. El modelo más sencillo de este tipo de caminata se ejecuta sobre un espacio discreto unidimensional (esto es, una recta con nodos). La evolución de esta caminata se lleva a cabo aplicando un operador de evolución consistente en dos operaciones cuánticas (dos operadores unitarios): la primera operación hace que la moneda entre en un estado cuántico que asemeja un volado, y la segunda operación hace que los componentes cuánticos de la moneda interactúen con el caminante, de tal suerte que la probabilidad de encontrar al caminante en distintos puntos de la línea sea una función del tiempo. La aplicación de las dos operaciones cuánticas es equivalente a un paso algorítmico, una operación elemental. La ecuación que define una caminata cuántica discreta es:
$$\mid{\psi_{n}}\rangle=\hat{U}^{n}\mid{\psi_{0}}\rangle$$

En el caso clásico, una caminata aleatoria en una cadena lineal se define sobre el conjunto de números enteros Z. Para una caminata cuántica, consideraremos un espacio de Hilbert definido sobre $Z$, es decir $H=l^{2}\{({\mid{x}\rangle:x\in{Z}})\}$ con los estados $\mid{x}\rangle$ forman una base ortonormal. En lugar de utilizar una matriz estocástica, podemos describir la evolución mediante una matriz unitaria. Esta simple "cuantificación" se ve como una traducción directa de caminatas al azar en el mundo cuántico. Al igual que para la caminata clásica, requerimos la invariancia traslacional de la evolución unitaria. Por lo tanto, si comenzamos en el estado $\mid{x-1}\rangle$, en el siguiente paso se convertirá en $\alpha\mid{x-2}\rangle+\beta\mid{x}\rangle$ para algunos (complejos) $\alpha$ y $\beta$. Similarmente el estado $\mid{x+1}\rangle$ se asignará al estado $\alpha\mid{x+1}\rangle+\beta\mid{x+2}\rangle$ con el mismo $\alpha$ y $\beta$ se tiene en cuenta que los dos estados inicialmente son ortogonales $\mid{x-1}\rangle$ y $\mid{x+1}\rangle$ debe permanecer ortogonal bajo cualquier evolución unitaria. Sin embargo, esto ahora es posible solo si uno de los coeficientes $\alpha$ o $\beta$ son cero. Agregamos otro grado de libertad, siendo este el espacio de monedas de tal forma que se puede describir la dirección del caminante. Este espacio adicional será un conjunto $\Xi=\{\uparrow,\downarrow\}$ para que el estado de una partícula sea descrito por una tupla $(x,c)\in{Z\times\Xi}$. Debido a esta adición, expandimos el espacio de Hilbert a un producto tensorial $H=H_{P}\otimes{H_{C}}$, donde $H_{P}=l^{2}(Z)$ determina una posición del caminante y $H_{C}=l^{2}(\Xi)$ es el espacio de monedas introducido. Por lo tanto $H$ se extiende por la base ortonormal $\{\mid{x}\rangle\otimes\mid{c}\rangle\:x\in{Z},C\in{\Xi}$. Un estado inicial $\mid{\psi_{0}}\rangle\in{H}$ evoluciona cuando se elige describir un solo paso de evolución por un operador de evolución unitaria, siendo este

$$U=S(I\otimes{C})$$

Es una composición de $I\otimes{C}$ actuando no trivialmente solo en el espacio de monedas, y $S$, que involucra a todo el espacio de Hilbert $H$. El operador $S$ describe la traducción espacial del caminante, mientras que $C$ corresponde al lanzamiento de monedas. Este operador actúa en el espacio de Hilbert $H$ como un operador de cambio de posición condicional con monedas, de tal forma que cambia la posición de la partícula a una posición hacia arriba si la moneda apunta hacia arriba y mueve la partícula a una posición hacia abajo, si la moneda apunta hacia abajo.

$$S=\sum_{x\in{Z}}(\mid{x+1}\rangle\langle{x}\mid\otimes\mid{\uparrow}\rangle\langle{\uparrow}\mid+\mid{x-1}\rangle\langle{x}\mid\otimes\mid{\downarrow}\rangle\langle{\downarrow}\mid)$$

El operador de monedas dado por $I\otimes{C}$ es un operador unitario que actúa solo en el espacio de monedas $H_{C}$ y corresponde a un "giro" de la moneda. Para la caminata cuántica en línea, $C$ es una matriz unitaria de $2\times{2}$. Una opción habitual para C es la moneda de Hadamard

$$C\equiv{H}=\frac{1}{\sqrt{2}}
\left(
\begin{array}{lcr}
1 & 1  \\
1 & -1  \\
\end{array}
\right)
$$

visto en una notación diferente como

$$H:\mid{\uparrow}\rangle\mapsto{\frac{1}{\sqrt{2}}(\mid{\uparrow}\rangle+\mid{\downarrow}\rangle)}$$

$$H:\mid{\downarrow}\rangle\mapsto{\frac{1}{\sqrt{2}}(\mid{\uparrow}\rangle-\mid{\downarrow}\rangle)}$$

Supongamos que el estado del sistema es $\mid{x}\rangle\mid{c}\rangle$ con $x\in{Z}$ y $c$ siendo uno de los estados base 
$\mid{\uparrow}\rangle$ y $\mid{\downarrow}\rangle$. Entonces, un solo paso de la caminata cuántica nos da

$$U\mid{x}\rangle\mid{c}\rangle=S(I\otimes{C})\mid{x}\rangle\mid{c}\rangle$$

$$=S(\mid{x}\rangle\otimes{H\mid{c}\rangle})$$

$$=\frac{1}{\sqrt{2}}S[\mid{x}\rangle\otimes(\mid{\uparrow}\rangle\pm\mid{\downarrow}\rangle)]$$

$$=\frac{1}{\sqrt{2}}(S\mid{x}\rangle\mid{\uparrow}\rangle\pm{S\mid{x}\rangle\mid{\downarrow}\rangle})$$

$$=\frac{1}{\sqrt{2}}\mid{x+1}\rangle\mid{\uparrow}\rangle\pm\mid{x-1}\rangle\mid{\downarrow}\rangle$$

De este resultado podemos observar que comenzando en el estado $\mid{\psi_{0}=\mid{x}\rangle\mid{\uparrow}\rangle}$ y realizar una medición en el espacio de estado de posición justo después de una aplicación de $U$, la probabilidad de encontrar $(x+1)$ es igual a la probabilidad de medir $(x-1)$. Por lo tanto, si realizamos una medición de posición después de cada aplicación de U, terminamos con una caminata aleatoria clásica donde cambiamos la posición a cualquier lado con igual probabilidad $\frac{1}{2}$.
Como no se quiere medir el sistema después de cada paso, entonces se puede observar que sucederá si se deja que el sistema evolucione durante algún tiempo y luego se mide la posición de la partícula. Con u ejemplo se puede ilustrar este caso con tres aplicaciones de $U$. Después de un desarrollo algebraico llegamos a

$$U^{3}\mid{0}\rangle\mid{\uparrow}\rangle=\frac{1}{\sqrt{8}}[\mid{3}\rangle\mid{\uparrow}\rangle+\mid{1}\rangle{(\mid{\uparrow}\rangle+2\mid{\downarrow}\rangle)}-\mid{-1}\rangle\mid{\uparrow}\rangle+\mid{-3}\rangle\mid{\downarrow}\rangle]$$

Esto ya muestra que las amplitudes no son simétricas y comienzan a mostrar efectos de interferencia. Después de m pasos, en la medición final, no nos importa el estado de la moneda. Por lo tanto, para obtener la probabilidad de que la partícula esté en la posición x después de m pasos, estará dada por

$$p^{m}(x)=\mid{(\mid{x}\rangle\otimes\mid{\uparrow}\rangle)U^{m}\mid{\psi_{0}}\rangle}\mid^{2}+\mid{(\mid{x}\rangle\otimes\mid{\downarrow}\rangle)U^{m}\mid{\psi_{0}}\rangle}\mid^{2}$$

La ejecución de varias caminatas cuánticas discretas, con estados iniciales idénticos y operaciones cuánticas iguales, permite generar distribuciones de probabilidad como las mostradas en las siguientes graficas  y estas a su vez ejemplifican algunas propiedades importantes de las caminatas cuánticas, las cuales son:

1.Las caminatas cuánticas discretas tienen una varianza que crece proporcionalmente al cuadrado del número de pasos, i.e.

$$\sigma_{q}^{2}(n)=O(n^{2})$$

Este hecho es importante por dos motivos: primero, la varianza de una caminata clásica es proporcional sólo al número de pasos ejecutados $((i.e.)\sigma_{q}^{2}(n)>\sigma^{2}(n))$ y segundo, esta diferencia entre las varianzas clásica y cuántica
puede ser utilizada para aumentar la velocidad de ejecución de un algoritmo basado en caminatas cuánticas, respecto del correspondiente algoritmo clásico diseñado con una caminata aleatoria.

2.La forma de la distribución de probabilidad generada con una caminata cuántica depende del estado inicial. Este hecho es importante pues el estado inicial del caminante y la moneda puede ser utilizado como un parámetro computacional.


```julia
#hacemos la siguiente rutina para obtener vectores de la base canónica de un espacio de cualquier dimensión.

function base(n,N)  #N es la dimensión del sistema y n el vector que se toma.
    a=zeros(N)
    a[n]=1
    return a
    end
```

```julia
#Rutina para construir el operador S de la expresión mostrada arriba

function S_operador(N) #N la dimensión
n=N  
    S=zeros(Float64,2*n,2*n)  #matriz de ceros cuya dimensión es 2N por el producto tensorial con los espinares con los elementos de la base canónica 
for i in 2:n-1
    S=S+kron(base(i+1,n).*base(i,n)',base(1,2)*base(1,2)')+kron(base(i-1,n)*base(i,n)',base(2,2).*base(2,2)') 
    
#iteramos la suma con todos los elementos de la base y guardamos en S.
        
end
    
 return S
    end 
```

Hacemos el operador U multiplicando la función S_operador por el producto tensorial entre la identidad y la matrz de Hadamard.

```julia
function u_operador(N)
S_operador(N)*(kron(eye(N),(1/sqrt(2))*[1 1 ; 1 -1])) 
end 
```

Ahora vamos a escribirir la rutina que nos de la probabilidad. 

```julia
function probabilidad(N,xo,m) #de nuevo N es la dimensión, xo el estado cuántico inicial y m las veces que vamos a aplicar el operdor U al estado.
U=u_operador(N)  
listP=Float64[]
for i in 1:N
   p= (kron(base(i,N)',base(1,2)')*(U^(m)*kron(base(xo,N),base(1,2)))).^2+(kron(base(i,N)',base(2,2)')*(U^(m)*kron(base(xo,N),base(1,2)))).^2
        push!(listP,p[1])
        
    end
    
        return listP
end
```

Ahora la meta es hacer una coomparación entre las densidades de probabilidad para la caminata clásica y cuántica.

Observemos que en el caso clásico el caminante siempre parte de la posición 0 y en el cuántico el estado inicial fue el vector canónico 50, entonces necesitamos hacer coincidir los puntos de partida para obtener una gráfica de las densidades de probabilidad satisfactoria.
Para lograrlo escribimos de nuevo las rutinas para el caso clásico usando la nueva condición inicial.

```julia
#La rutina que se presenta a continuación es una recopilación  de diversas funciones. El input (n) corresponde al número de volados que se desea realizar.
function posicion_final2(n)
    b=rand(-1:2:1,n)     #Usamos la función rand para generar valores aleatorios, éstos sólo podrán ser 1 (sol) y -1 (águila).
    p=50
    for i in 1:length(b)
    p=b[i]+p  #Se implementa un for para, una vez generados todos los valores de los volados, obtener la posición final del caminante, esto es simplemente la suma de todos ellos.
end
return p  #Finalmente, el valor p corresponde a la posición final del caminante después de n pasos.
end
```

```julia
#Sin embargo, a nosotros nos interesa hacer un análisis para  K número de caminantes. Ahora le pedimos al usuario tanto el número de volados "n", como el de caminantes "k"
function pf_caminantes2(n,k) 
    listf=zeros(k)  # Construimos una lista con k entradas de ceros 
    for i in 1:k  #Este ciclo for se usa para encontrar la posición final de los k caminantes haciendo uso de la función de posicion_final (presentada anteriormente)
        d=posicion_final2(n)
        listf[i]=d  
        end 
    return listf # Se imprime la lista con la posición final de los k caminantes.
    end
```

Hasta el momento vimos algunas de las diferencias que presenta el caso clásico al compararlo con el cuántico; estudiamos la distribución de probabilidad del caso clásico para una condición inicial localizada y encontramos un comportamiendo muy diferente al caso cuántico (como se muestra en la gráfica previa). La representanción del caso clásico, como ya se ha mencionado, corresponde a un comportamiento que da como resultado una gráfica acampanada y simétrica respecto al punto de partida, también llamada campana de gauss. la forma de esta distribción binomial no depende del punto de partida. Lo que sucede al cambiar el punto inicial, es que la gráfica se desplazará a la izquierda o a la derecha, pero la forma será la misma. Esta invariancia de la forma de la distribución respecto del punto de partida es una característica fundamental de las cadenas de Markov, de las cuales las caminatas aleatorias son un caso especial.
Mientras que en el segundo caso, La forma de la distribución de probabilidad generada con una caminata cuántica depende del estado inicial. Para este caso ocurre que en cada paso la partícula se coloca en una superposición cuántica de ir de un modo y otro. Cuando los diversos caminos se encuentran de nuevo, se produce interferencia y la distribución de probabilidades de encontrar la partícula en ubicaciones determinadas difiere mucho de la situación clásica. En particular, la probabilidad de encontrar la partícula en varias ubicaciones se extiende mucho más rápida y exhaustivamente a través del espacio que en el caso clásico.

## Anexo

Demostración de que S es un operador unitario. 

$$S=\mid{\uparrow}\rangle\langle{\uparrow}\mid\otimes\sum_{x\in{Z}}\mid{x+1}\rangle\langle{x}\mid +\mid{\downarrow}\rangle\langle{\downarrow}\mid\otimes\sum_{x\in{Z}}\mid{x-1}\rangle\langle{x}\mid)$$

$$S^{\dagger}=(\mid{\uparrow}\rangle\langle{\uparrow}\mid\otimes\sum_{x\in{Z}}\mid{x+1}\rangle\langle{x}\mid +\mid{\downarrow}\rangle\langle{\downarrow}\mid\otimes\sum_{x\in{Z}}\mid{x-1}\rangle\langle{x}\mid)^{\dagger}$$

Por las propiedades del operador adjunto tenemos:

$$S^{\dagger}=(\mid{\uparrow}\rangle\langle{\uparrow}\mid)^{\dagger}\otimes\sum_{x\in{Z}}(\mid{x+1}\rangle\langle{x}\mid)^{\dagger} +(\mid{\downarrow}\rangle\langle{\downarrow}\mid)^{\dagger}\otimes\sum_{x\in{Z}}(\mid{x-1}\rangle\langle{x}\mid)^{\dagger}$$

Al multiplicar $S$ y $S^{\dagger}$

$$SS^{\dagger}=(\mid{\uparrow}\rangle\langle{\uparrow}\mid\otimes\sum_{x\in{Z}}\mid{x+1}\rangle\langle{x}\mid)((\mid{\uparrow}\rangle\langle{\uparrow}\mid)^{\dagger}\otimes\sum_{x\in{Z}}(\mid{x+1}\rangle\langle{x}\mid)^{\dagger})+(\mid{\uparrow}\rangle\langle{\uparrow}\mid\otimes\sum_{x\in{Z}}\mid{x+1}\rangle\langle{x}\mid)((\mid{\downarrow}\rangle\langle{\downarrow}\mid)^{\dagger}\otimes\sum_{x\in{Z}}(\mid{x-1}\rangle\langle{x}\mid)^{\dagger})+(\mid{\downarrow}\rangle\langle{\downarrow}\mid\otimes\sum_{x\in{Z}}\mid{x-1}\rangle\langle{x}\mid)((\mid{\uparrow}\rangle\langle{\uparrow}\mid)^{\dagger}\otimes\sum_{x\in{Z}}(\mid{x+1}\rangle\langle{x}\mid)^{\dagger})+(\mid{\downarrow}\rangle\langle{\downarrow}\mid\otimes\sum_{x\in{Z}}\mid{x-1}\rangle\langle{x}\mid)((\mid{\downarrow}\rangle\langle{\downarrow}\mid)^{\dagger}\otimes\sum_{x\in{Z}}(\mid{x-1}\rangle\langle{x}\mid)^{\dagger}) $$

Usamos la siguiente propiedad de los productos tensoriales:  $(A\otimes B)(C\otimes D)=AC\otimes BD$

$$SS^{\dagger}=(\mid{\uparrow}\rangle\langle{\uparrow}\mid)(\mid{\uparrow}\rangle\langle{\uparrow}\mid)^{\dagger}\otimes\sum_{x\in{Z}}\mid{x+1}\rangle\langle{x}\mid(\mid{x+1}\rangle\langle{x}\mid)^{\dagger}+(\mid{\uparrow}\rangle\langle{\uparrow}\mid)(\mid{\downarrow}\rangle\langle{\downarrow}\mid)^{\dagger}\otimes\sum_{x\in{Z}}\mid{x+1}\rangle\langle{x}\mid(\mid{x-1}\rangle\langle{x}\mid)^{\dagger}+(\mid{\downarrow}\rangle\langle{\downarrow}\mid)(\mid{\uparrow}\rangle\langle{\uparrow}\mid)^{\dagger}\otimes\sum_{x\in{Z}}\mid{x-1}\rangle\langle{x}\mid(\mid{x+1}\rangle\langle{x}\mid)^{\dagger}+(\mid{\downarrow}\rangle\langle{\downarrow}\mid)(\mid{\downarrow}\rangle\langle{\downarrow}\mid)^{\dagger}\otimes\sum_{x\in{Z}}\mid{x-1}\rangle\langle{x}\mid(\mid{x-1}\rangle\langle{x}\mid)^{\dagger}$$

Los términos cruzados son cero debido a que 
$$\mid{\uparrow}\rangle\langle{\uparrow}\mid=
\left(
\begin{array}{lcr}
1 & 0  \\
0 & 0  \\
\end{array}
\right)
$$

$$\mid{\downarrow}\rangle\langle{\downarrow}\mid=
\left(
\begin{array}{lcr}
0 & 0  \\
0 & 1  \\
\end{array}
\right)
$$

Y además son iguales a su adjunto, es decir. Son hermitianos

Por tanto el producto es:

$$SS^{\dagger}=(\mid{\uparrow}\rangle\langle{\uparrow}\mid)\otimes\sum_{x\in{Z}}\mid{x+1}\rangle\langle{x}\mid(\mid{x+1}\rangle\langle{x}\mid)^{\dagger}+(\mid{\downarrow}\rangle\langle{\downarrow}\mid)\otimes\sum_{x\in{Z}}\mid{x-1}\rangle\langle{x}\mid(\mid{x-1}\rangle\langle{x}\mid)^{\dagger}$$

Aplicando que el adjunto de un ket es un bra obtenemos:

$$SS^{\dagger}=(\mid{\uparrow}\rangle\langle{\uparrow}\mid)\otimes\sum_{x\in{Z}}\mid{x+1}\rangle\langle{x+1}\mid\langle{x} \mid{x}\rangle+(\mid{\downarrow}\rangle\langle{\downarrow}\mid)\otimes\sum_{x\in{Z}}\mid{x-1}\rangle\langle{x-1}\mid\langle{x}\mid{x}\rangle$$

Usamos que:
$$\langle{x}\mid{x}\rangle=1$$

y 
$$\mid{\uparrow}\rangle\langle{\uparrow}\mid)+\mid{\downarrow}\rangle\langle{\downarrow}\mid=\mathcal{I}_{2}$$


Donde $\mathcal{I}_{2}$ es la matriz identidad de 2-D$

$$SS^{\dagger}=\mathcal{I}_{2}\otimes\sum_{x\in{Z}}\mid{x+1}\rangle\langle{x+1}\mid-(\mid{\downarrow}\rangle\langle{\downarrow}\mid)\otimes\sum_{x\in{Z}}\mid{x+1}\rangle\langle{x+1}\mid+(\mid{\downarrow}\rangle\langle{\downarrow}\mid)\otimes\sum_{x\in{Z}}\mid{x-1}\rangle\langle{x-1}\mid$$

Y tomando en cuenta que la matriz identidad del espacio n dimensional es.

$$\sum_{x\in{Z}}\mid{x}\rangle\langle{x}\mid=\mathcal{I}_{n}$$

Los ultimos dos términos de la expresión del producto se cancelan y tenemos la expesión: 


$$SS^{\dagger}=\mathcal{I}_{2}\otimes\mathcal{I}_{n}$$

Por tanto se tiene que: 
$$SS^{\dagger}=\mathcal{I}_{2n}$$

Y al satisfacer lo anterior concluimos que $S$ es un operador unitario.
