# Ejercicio 1

Implementar método implicito de Euler como quieran.

```julia
#Se deine la función de dos variables a resolver por los diversos métodos.
function f(x,t)
    return -2.5x
end
```

```julia
#Se define función para la solución exacta, ésta sólo depende de t.
function solucion(t)
    return e^(-2.5*t)
end
```

```julia
#Se define la función método de Newton como lo hicimos en la tarea 4, sin embargo, ésta nueva varía en las condiciones iniciales.
function metodo_Newton(f,x0,t)
    z=x0
    h=[]
    df=diff(f(x,t),x)
    for i in 1:100
    z=z-(f(z,t)/df(z,t));
        push!(h,z)
    end
    return z
end
```

```julia
#Ahora se define la función: euler implícito, básicamente el código corresponde al del método de Euler (ver tarea 11), las modificaciones son, el definir una nueva función para obtener las raíces, y aplicar directamente el método de Newton.
function Euler_implicito(f,x0,t0,t,h)
    listt=t0:h:t
    listx=[x0]
    for i in 1:length(listt)-1
         g(x)=x-x0-h*f(x,listt[i])
         x0=x0+h*f(metodo_Newton(g,Float64(x0),listt[i]),listt[i+1])
        push!(listx,x0)
    end
    return listt,listx
end
```

```julia
#Ahora se considera la primera condición inicial x(0)=1, se toman los límites del intervalo [0,8] y finalmente se varía el valor de h, según cual sea el caso.
EI2=Euler_implicito(f,1.0,0.0,8.0,0.5)
EI3=Euler_implicito(f,1.0,0.0,8.0,0.2)
EI4=Euler_implicito(f,1.0,0.0,8.0,0.01)
```

```julia
EI2[1]
```

```julia
EI2[1]
```

```julia
x=linspace(0,0.1,10)
title("Método de Euler Implícito")
plot(EI2[1],EI2[2],label="h=0.5")
plot(EI3[1],EI3[2],label="h=0.2")
plot(EI4[1],EI4[2],label="h=0.01")
plot(x,solucion.(x),label="solución exacta")
legend()
```

# Ejercicio 2

Implementar Euler: Regla del punto medio.

```julia
#Para definir la función de punto medio, nuevamente se utiliza la estructura de la función de Euler (tarea 11), simplemente se modifica la aproximación al punto medio n+1/2.
function metodo_puntomedio(f,x0,t0,t,h)
    listt=t0:h:t
    listx=[x0]
    x=x0
    for i in 1:length(listt)-1
        x=x+h*f(x+(h/2)f(x,listt[i]),listt[i]+h/2)
        push!(listx,x)
    end
    return listt,listx
end
```

```julia
PM1=metodo_puntomedio(f,1.0,0.0,8.0,1.0)
PM2=metodo_puntomedio(f,1.0,0.0,8.0,0.5)
PM3=metodo_puntomedio(f,1.0,0.0,8.0,0.2)
PM4=metodo_puntomedio(f,1.0,0.0,8.0,0.01) 
```

```julia
PM4[2]
```

```julia
x=linspace(0,0.1,10)
title("Método del punto medio")
plot(PM2[1],PM2[2],label="h=0.5")
plot(PM3[1],PM3[2],label="h=0.2")
plot(PM4[1],PM4[2],label="h=0.01")
plot(x,solucion.(x),label="solución exacta")
legend()
```

# Ejercicio 3 

Implementar método de Runge-Kutta de orden 4.

```julia
#Una vez más se utiliza el método de Euler, ahora se definen coeficientes ki (el número de coeficientes cambiará según cual sea el orden) para aproximaciones intermedias.
function runge_kutta(f,x0,t0,t,h)
    listt=t0:h:t
    listx=[x0]
    x=x0
    for i in 1:length(listt)-1
        k1=f(x,listt[i])
        k2=f(x+(h/2)*k1,listt[i]+h/2)
        k3=f(x+(h/2)*k2,listt[i]+h/2)
        k4=f(x+h*k3,listt[i]+h)
        x=x+(h/6)*(k1+2*k2+2*k3+k4)
   push!(listx,x)
    end
    return listt,listx
end
```

```julia
RK2=runge_kutta(f,1.0,0.0,8.0,0.5)
RK3=runge_kutta(f,1.0,0.0,8.0,0.2)
RK4=runge_kutta(f,1.0,0.0,8.0,0.01)
```

```julia
x=linspace(0,0.1,10)
title("Método de Runge Kutta de orden 4")
plot(RK2[1],RK2[2],label="h=0.5")
plot(RK3[1],RK3[2],label="h=0.2")
plot(RK4[1],RK4[2],label="h=0.01")
plot(x,solucion.(x),label="solución exacta")
legend()
```

```julia
#Euler explícito es tal cual el código de Euler que fue presentado en la tarea pasada.
function euler_explicito(f,x0,t0,t,h)
    listt=t0:h:t
    listx=[x0]
    for i in 1:length(listt)-1
        x0=x0+f(Float64(x0),listt[i])*h
        push!(listx,x0)
    end
    return listt,listx
end
```

```julia
EE2=euler_explicito(f,1.0,0.0,8.0,0.5)
EE3=euler_explicito(f,1.0,0.0,8.0,0.2)
EE4=euler_explicito(f,1.0,0.0,8.0,0.01)
```

```julia
x=linspace(0,0.1,10)
title("Método de Euler Explícito")
plot(EE2[1],EE2[2],label="h=0.5")
plot(EE3[1],EE3[2],label="h=0.2")
plot(EE4[1],EE4[2],label="h=0.01")
plot(x,solucion.(x),label="solución exacta")
legend()
```

Haciendo un comparativo entre el método de Euler explícito e implícito, nos resulta claro apreciar que éste último se aproxima con mayor exactitud a la solución de la ecuación diferencial (véase ejercicio 1), esto es, el método implícito de Euler arroja una solución más precisa.

# Ejercicio 6

Grafique la energía total del oscilador armónico usando los métodos explícito de Euler y Runge-Kutta de orden 4. Compare sus respuestas y discuta.

```julia
#Se define la ecuación diferencial de dos dimensiones a resolver.
function D(x,t)
    return [x[2], -1.0*x[1]]
end
```

```julia
#A continuación se presenta el código para calcular una ecuación diferencial por el método de Euler, éste a comparación del primer ejercicio del método de Euler , es para sistemas de más de una dimensión.
function euler(f,list,x0)
     x = x0
     h = list[2]-list[1]
     listx=[]
     push!(listx,x)
     for i in 2:length(list)
        t = i*h
        x = x + f(x,t)*h
        push!(listx,x) 
     end
     return listx
end
```

```julia
#Se define el intervalo donde se hará la representación gráfica, y un vector denominado "condiciones iniciales" que contiene las soluciones de nuestro sistema.
list=0:0.01:6;
condiciones_iniciales=euler(D,list,[0,1]);
```

```julia
listx1=map(x->x[1],condiciones_iniciales);
listx2=map(x->x[2],condiciones_iniciales);
plot(list,listx1)
plot(list,listx2)
legend()
```

```julia
#Nuevamente se presenta el código para resolver una ecuación diferencial de más de una dimensión, pero ahora por el método de Runge Kutta.
function runge_kutta2(f,list,x0)
     x = x0
     h = list[2]-list[1]
     listx=[]
     push!(listx,x)
     for i in 2:length(list)
        t = i*h
        k1 = f(x,t);
        k2 = f(x+(h/2)*k1,t+(h/2));
        k3 = f(x+(h/2)*k2, t+(h/2));
        k4 = f(x+h*k3, t+h);
        x = x+(h/6)*(k1+2*k2+2*k3+k4);
        push!(listx,x) 
     end
     return listx
end
```

```julia
condiciones_iniciales2=runge_kutta2(D,list,[0,1]);
```

```julia
listx3=map(x->x[1],condiciones_iniciales2);
listx4=map(x->x[2],condiciones_iniciales2);
plot(list,listx3)
plot(list,listx4)
legend()
```

```julia
#Ahora se crean las funciones de energía, energía potencial y energía cinética.
function energia(x,y)
    return  ((x^2)+(y^2))/2
end
```

```julia
function potencial(x,y)
    return  (y^2)/2
end
```

```julia
function cinetica(x,y)
    return  (x^2)/2
end
```

```julia
#Se grafican las tres energías para el método de Euler.
x=linspace(0,0.01,10)
title("Energía por el método de Euler")
plot(list,energia_total.(listx1,listx2), label="Energía total");
plot(list,potencial.(listx1,listx2));
plot(list,cinetica.(listx1,listx2));
legend();
```

```julia
#Se grafican las tres energías para el método de Runge Kutta.
x=linspace(0,0.01,10)
title("Energía por el método de Runge Kutta")
plot(list,energia_total.(listx3,listx4), label="Energía total");
plot(list,potencial.(listx3,listx4));
plot(list,cinetica.(listx3,listx4));
legend();
```
