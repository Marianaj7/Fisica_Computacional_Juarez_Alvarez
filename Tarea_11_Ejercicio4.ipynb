# Ejercicio 4
Obten la solución del sistema $\ddot{x} = -kx(t)$ utilizando lo escrito arriba y el método de Euler.
Para utilizar el método de Euler, impleméntelo para sistemas de más de una dimensión.

```julia
#Se define la ecuación diferencial de dos dimensiones a resolver por medio del método de Euler. 
function D(x,t)
    return [x[2], -1.0*x[1]]
end
```

```julia
#A continuación se presenta el código para calcular una ecuación diferencial por el método de Euler, éste a comparación del primer ejercicio , es para sistemas de más de una dimensión.  
function euler(f,list,x0)
     x = x0
     h = list[2]-list[1]
     listx=[]
     push!(listx,x)
     for i in 2:length(list)
        t = i*h
        x = x + f(x,t)*h
        push!(listx,x) 
     end
     return listx
end
```

```julia
#Se define el intervalo donde se hará la representación gráfica, y un vector denominado "condiciones iniciales" que contiene las soluciones de nuestro sistema, la primera entrada es una condición inicial.
list=0:0.01:10;
condiciones_iniciales=euler(D,list,[0,2]);
```

```julia
listx1=map(x->x[1],condiciones_iniciales);
listx2=map(x->x[2],condiciones_iniciales);
```

```julia
plot!(list,listx1)
plot!(list,listx2)
```
