# Diagrama de bifurcación

Un diagrama de bifurcación de un sistema dinámico es una estratificación de su espacio de parámetros inducida por la equivalencia topológica, junto con los retratos de fase representativos de cada estrato. Las soluciones estables suelen representarse mediante líneas continuas, mientras que las soluciones inestables se representan con líneas punteadas.

$$ x_{n+1}=rx_n(1- x_n), r \epsilon [0,4]$$ y $$x\epsilon [0,1]$$

```julia
#Se instala la librería a utilizar durante todo el notebook.
using PyPlot
```

```julia
#A continuación se presenta el programa con el que podremos obtener nuestro diagrama de difurcación, se implementa un for para hacer diversas iteraciones.
function diagrama_bifurcacion(x0,r,n)
        for i in 1:n
        x0=r*x0*(1-x0)
        end
    return x0
end
```

```julia
#Ahora definimos un par de intervalos linspace en los que estarán definidos los valores de x y r, nuevamente se implementa un for para hacer diversas iteraciones y nos sea posible graficar. 
x=linspace(0,1,1000)
r=linspace(0,4,1000)
D=[]
B=[]
for i in 1:length(r)
    push!(D,diagrama_bifurcacion.(x,r[i],100))
    push!(B,ones(length(x))*r[i])
end
```

```julia
#Se grafica el diagrama de bifurcación.
plot(B,D,alpha=0.1,"r.",markersize=0.5)
title("Diagrama de bifurcación")
axes()[:set_xlim]([1,4])
axes()[:set_ylim]([0,1])
```

# Mapeo Logístico y Cálculo del Exponente de Lyapunov


```julia
#Asimismo se presenta el programa para calcular el exponente de Lyapunov.
function exp_lyapunov(x,n,r)
    l=0    
for i in 1:n                                 
    x = r*x*(1-x)                            
    l = l+log(abs(r*(1-2*x)))                 
    end
    return l/n
    end
```

```julia
#Finalmente graficamos.
plot(B,exp_lyapunov.(0.5,1000,r),"r.",alpha=0.05,markersize=0.5)
title("Exponente de Lyapunov")
axes()[:set_xlim]([1,4])
axes()[:set_ylim]([-6,1])
```
