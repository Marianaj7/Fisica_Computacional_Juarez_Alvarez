# Euler

```julia
D(x,t)=[x[2], -1.0*x[1]-1.0*x[2]]
list=0:0.01:100;
EE6=herramientas.euler(D,list,[0,1]);
list1=map(x->x[1],EE6);
list2=map(x->x[2],EE6);
plot(list1,list2)
```

```julia
h(x,t)=x^2
EI2=herramientas.Euler_implicito(h,1.0,0.0,10.0,0.1)
EI3=herramientas.Euler_implicito(h,1.0,0.0,10.0,0.001)
plot(EI2[1],EI2[2])
plot(EI3[1],EI3[2])
axes()[:set_xlim]([0,20])
axes()[:set_ylim]([0,20])
```

# Runge Kutta

```julia
D(x,t)=[x[2], -1.0*x[1]-1.0*x[2]]
list=0:0.01:100;
RK7=herramientas.runge_kutta2(D,list,[0,1]);
list3=map(x->x[1],RK7);
list4=map(x->x[2],RK7);
plot(list3,list4)
```
